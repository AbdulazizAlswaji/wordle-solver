```python
words = open('wordle.txt', 'r')
lines = words.readlines()
words.close()
```

```python
wordle = pd.DataFrame(columns={
    'word'
})
wordle.word = lines
wordle.word = wordle.word.map(lambda w: w.replace('\n', ''))
wordle = wordle[wordle.word != '']
wordle = wordle.sort_values('word').reset_index()
del wordle['index']

wordle['w0'] = wordle.word.map(lambda w: w[0:1])
wordle['w1'] = wordle.word.map(lambda w: w[1:2])
wordle['w2'] = wordle.word.map(lambda w: w[2:3])
wordle['w3'] = wordle.word.map(lambda w: w[3:4])
wordle['w4'] = wordle.word.map(lambda w: w[4:])

wordle.head(10)
```

```python
common = wordle[(wordle.word.str.contains('e')) & (wordle.word.str.contains('o'))].reset_index()
i = randrange(len(common) - 1)
first_attempt = common[common.index == i].reset_index()

w0 = first_attempt.iloc[0].w0.upper()
w1 = first_attempt.iloc[0].w1.upper()
w2 = first_attempt.iloc[0].w2.upper()
w3 = first_attempt.iloc[0].w3.upper()
w4 = first_attempt.iloc[0].w4.upper()

driver.execute_script(letters[w0]).click()
driver.execute_script(letters[w1]).click()
driver.execute_script(letters[w2]).click()
driver.execute_script(letters[w3]).click()
driver.execute_script(letters[w4]).click()
driver.execute_script(letters['ENTER']).click()
    

first_attempt
```
